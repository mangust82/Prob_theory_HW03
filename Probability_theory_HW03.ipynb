```python
# –î–∞–Ω—ã –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞—Ä–ø–ª–∞—Ç –∏–∑ –≤—ã–±–æ—Ä–∫–∏ –≤—ã–ø—É—Å–∫–Ω–∏–∫–æ–≤: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150.
# –ü–æ—Å—á–∏—Ç–∞—Ç—å (–∂–µ–ª–∞—Ç–µ–ª—å–Ω–æ –±–µ–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö –º–µ—Ç–æ–¥–æ–≤ –Ω–∞–ø–æ–¥–æ–±–∏–µ std, var, mean) —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ, 
# —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ, —Å–º–µ—â–µ–Ω–Ω—É—é –∏ –Ω–µ—Å–º–µ—â–µ–Ω–Ω—É—é –æ—Ü–µ–Ω–∫–∏ –¥–∏—Å–ø–µ—Ä—Å–∏–π –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏.
import pandas as pd

salary_str = '100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150'
salary = list(map(int, salary_str.split(', ')))
salary.sort()
print(salary)
# —Å—Ä–µ–¥–Ω–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ
mean = sum(salary)/len(salary)
print(f'—Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ {mean}')
# –Ω–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏
sumX = 0
for i in salary:
    X = pow(i - mean, 2)
    sumX += X
Dipersia_non = sumX/(len(salary)-1)
print(f'–Ω–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ {Dipersia_non}')
# —Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏
Dipersia = sumX/(len(salary))
print(f'—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ {Dipersia}')
# —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
STD = pow(Dipersia_non, 0.5)
print(f'—Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ {STD}')

df = pd.DataFrame(salary)
print()
# print(df)
mean = df.mean()
print(mean)
Dipersia = df.var(ddof=0)
print(Dipersia)
```

```python

# –í –ø–µ—Ä–≤–æ–º —è—â–∏–∫–µ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è 8 –º—è—á–µ–π, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö 5 - –±–µ–ª—ã–µ. –í–æ –≤—Ç–æ—Ä–æ–º —è—â–∏–∫–µ - 12 –º—è—á–µ–π, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö 5 –±–µ–ª—ã—Ö. –ò–∑ –ø–µ—Ä–≤–æ–≥–æ —è—â–∏–∫–∞ 
# –≤—ã—Ç–∞—Å–∫–∏–≤–∞—é—Ç —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º –¥–≤–∞ –º—è—á–∞, –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ - 4. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ 3 –º—è—á–∞ –±–µ–ª—ã–µ?

from IPython.display import display
from numpy import math

def composite(n, k):
    N = math.factorial(n)/(math.factorial(k)*math.factorial(n-k))
    return N
# –î–ª—è —Ä–µ—à–µ–Ω–∏—è —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã—á–∏—Å–ª–∏—Ç—å –ø–æ–ª–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö —Å–æ–±—ã—Ç–∏–π. –ê –∏–º–µ–Ω–Ω–æ:
#     –∑–¥–µ—Å—å –≤–æ–∑–º–æ–∂–Ω–∞ –∫–æ–º–±–∏–Ω–∞—Ü–∏—è 1 —è—â–∏–∫ - 0 –±–µ–ª—ã—Ö —à–∞—Ä–æ–≤, 2 —è—â–∏–∫ 3 –±–µ–ª—ã—Ö —à–∞—Ä–∞; 1 —è—â–∏–∫ - 1 –±–µ–ª—ã–π —à–∞—Ä, 2 —è—â–∏–∫ - 2 –±–µ–ª—ã—Ö —à–∞—Ä–∞;
#     1 —è—â–∏–∫ - 2 –±–µ–ª—ã—Ö —à–∞—Ä–æ–≤, 2 —è—â–∏–∫ - 1 –±–µ–ª—ã–π —à–∞—Ä
# –í—ã—á–∏—Å–ª–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Å–∫–æ–º—ã—Ö –∏—Å—Ö–æ–¥–æ–≤ –¥–ª—è 1-–≥–æ –∏ 2-–≥–æ —è—â–∏–∫–∞
print('–í—ã—á–∏—Å–ª–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Å–∫–æ–º—ã—Ö –∏—Å—Ö–æ–¥–æ–≤ –¥–ª—è 1-–≥–æ –∏ 2-–≥–æ —è—â–∏–∫–∞\n box1M1w = composite(5, 0)\n box1M1b = composite(3, 2)...')
box1M1w = composite(5, 0)
box1M1b = composite(3, 2)
box1M2w = composite(5, 1)
box1M2b = composite(3, 1)
box1M3w = composite(5, 2)
box1M3b = composite(3, 0)
box1N = composite(8, 2)

box2M1w = composite(5, 3)
box2M1b = composite(7, 1)
box2M2w = composite(5, 2)
box2M2b = composite(7, 2)
box2M3w = composite(5, 1)
box2M3b = composite(7, 3)
box2N = composite(12, 4)

# –¢–æ–≥–¥–∞ –ø–æ–ª–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å: 
print('–¢–æ–≥–¥–∞ –ø–æ–ª–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å: Pcom = (box1M1w*box1M1b/box1N)*(box2M1w*box2M1b/box2N) + (box1M2w*box1M2b/box1N)*(box2M2w*box2M2b/box2N) \
+ (box1M3w*box1M3b/box1N)*(box2M3w*box2M3b/box2N)')
Pcom = (box1M1w*box1M1b/box1N)*(box2M1w*box2M1b/box2N) + (box1M2w*box1M2b/box1N)*(box2M2w*box2M2b/box2N) \
+ (box1M3w*box1M3b/box1N)*(box2M3w*box2M3b/box2N)
print(f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ —á—Ç–æ –≤—Å–µ 3 –º—è—á–∞  –∏–∑ 6 –±–µ–ª—ã–µ {Pcom}\n')

# –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–π —Å–ø–æ—Å–æ–±:
# –ü–æ—Å–∫–æ–ª—å–∫—É —Å–æ–±—ã—Ç–∏–µ –¥–æ—Å—Ç–∞—Ç—å —à–∞—Ä —Ä–∞–≤–Ω–æ–≤–µ—Ä–æ—è—Ç–Ω–æ –¥–ª—è –ª—é–±–æ–π –∫–æ—Ä–∑–∏–Ω—ã —Å–¥–ª–µ–∞–µ–º –¥–æ–ø—É—â–µ–Ω–∏–µ –∏ —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –∫–∞–∫ –±—É–¥—Ç–æ —à–∞—Ä—ã –ª–µ–∂–∞—Ç –≤ –æ–¥–Ω–æ–π –∫–æ—Ä–∑–∏–Ω–µ
# –¢–æ–≥–¥–∞ —É—Å–ª–æ–≤–∏–µ –º–æ–∂–Ω–æ –ø–µ—Ä–µ—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å —Å–ª–µ–¥ –æ–±—Ä–∞–∑–æ–º –≤ –∫–æ—Ä–∑–∏–Ω–µ –ª–µ–∂–∏—Ç 20 —à–∞—Ä–æ–≤ –∏–∑ –Ω–∏ 10 –±–µ–ª—ã—Ö 10 —á–µ—Ä–Ω—ã—Ö. –¢–æ–≥–¥–∞ –Ω–∞–π–¥–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ 
# —á—Ç–æ 3 —à–∞—Ä–∞ –∏–∑ 6 –±–µ–ª—ã–µ:
print('–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–π —Å–ø–æ—Å–æ–±:')
print('–ü–æ—Å–∫–æ–ª—å–∫—É —Å–æ–±—ã—Ç–∏–µ –¥–æ—Å—Ç–∞—Ç—å —à–∞—Ä —Ä–∞–≤–Ω–æ–≤–µ—Ä–æ—è—Ç–Ω–æ –¥–ª—è –ª—é–±–æ–π –∫–æ—Ä–∑–∏–Ω—ã —Å–¥–µ–ª–µ–∞–µ–º –¥–æ–ø—É—â–µ–Ω–∏–µ –∏ —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –∫–∞–∫ –±—É–¥—Ç–æ —à–∞—Ä—ã –ª–µ–∂–∞—Ç \
–≤ –æ–¥–Ω–æ–π –∫–æ—Ä–∑–∏–Ω–µ –¢–æ–≥–¥–∞ —É—Å–ª–æ–≤–∏–µ –º–æ–∂–Ω–æ –ø–µ—Ä–µ—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å —Å–ª–µ–¥ –æ–±—Ä–∞–∑–æ–º –≤ –∫–æ—Ä–∑–∏–Ω–µ –ª–µ–∂–∏—Ç 20 —à–∞—Ä–æ–≤ –∏–∑ –Ω–∏ 10 –±–µ–ª—ã—Ö 10 —á–µ—Ä–Ω—ã—Ö. \
–¢–æ–≥–¥–∞ –Ω–∞–π–¥–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ —á—Ç–æ 3 —à–∞—Ä–∞ –∏–∑ 6 –±–µ–ª—ã–µ')

Ncom = composite(20, 6)
Mwcom = composite(10, 3)
Mbcom = composite(10, 3)
Pw = (Mwcom * Mbcom) / Ncom
print(f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ —á—Ç–æ –≤—Å–µ 3 –º—è—á–∞ –∏–∑ 6 –±–µ–ª—ã–µ  Pw = N/M = {Pw}\n')

```

```python
# –ù–∞ —Å–æ—Ä–µ–≤–Ω–æ–≤–∞–Ω–∏–∏ –ø–æ –±–∏–∞—Ç–ª–æ–Ω—É –æ–¥–∏–Ω –∏–∑ —Ç—Ä–µ—Ö —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–≤ —Å—Ç—Ä–µ–ª—è–µ—Ç –∏ –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –º–∏—à–µ–Ω—å. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è –¥–ª—è –ø–µ—Ä–≤–æ–≥–æ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–∞ 
# —Ä–∞–≤–Ω–∞ 0.9, –¥–ª—è –≤—Ç–æ—Ä–æ–≥–æ ‚Äî 0.8, –¥–ª—è —Ç—Ä–µ—Ç—å–µ–≥–æ ‚Äî 0.6. –ù–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω: a). –ø–µ—Ä–≤—ã–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º 
# –±). –≤—Ç–æ—Ä—ã–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º –≤). —Ç—Ä–µ—Ç—å–∏–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º.

print('–ó–∞–ø–∏—à–µ–º —Ñ–æ—Ä–º—É–ª—É –ë–∞–π–µ—Å–∞ –¥–ª—è –ø—Ä–æ–∏–∑–æ—à–µ–¥—à–µ–≥–æ —Å–æ–±—ã—Ç–∏—è: P(B|A) = P(A|B)*P(B)/P(A), –≥–¥–µ P(A|B) –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è —Ç–æ–≥–æ –∏–ª–∏ \
–∏–Ω–æ–≥–æ —Å—Ç—Ä–µ–ª–∫–∞ –µ—Å–ª–∏ —Å—Ç—Ä–µ–ª—è–ª –æ–Ω(—Å–æ–±—ã—Ç–∏–µ –í), –≥–¥–µ P(A|B) –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è –±–∏–∞—Ç–ª–æ–Ω–∏—Å—Ç–æ–º, –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ —á—Ç–æ —Å—Ç—Ä–µ–ª—è–ª –æ–Ω. \
P(B)- –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ —á—Ç–æ —Å—Ç—Ä–µ–ª—è–ª –±–∏–∞—Ç–ª–æ–Ω–∏—Å—Ç —Ä–∞–≤–Ω–∞ 1/3 —Å–æ–±—ã—Ç–∏–µ —Ä–∞–≤–Ω–æ–≤–µ—Ä–æ—è—Ç–Ω–æ. P(A) - –ø–æ–ª–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å. ')
print('–í—ã—á–∏—Å–ª–∏–º –ø–æ–ª–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(A) = P(ùêµ_1)‚àóùëÉ(ùê¥‚îÇùêµ_1 )+ùëÉ(ùêµ_2 )‚àóùëÉ(ùê¥‚îÇùêµ_2 )+ùëÉ(ùêµ_3 )‚àóùëÉ(ùê¥‚îÇùêµ_3 )')
Pa = (1/3)*0.9+(1/3)*0.8+(1/3)*0.6
print(f'Pa = {Pa}')
print('–í—ã—á–∏—Å–ª–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —á—Ç–æ —ç—Ç–æ –±—ã–ª –ø–µ—Ä–≤—ã–π –±–∏–∞—Ç–ª–æ–Ω–∏—Å—Ç P(B|A)1')
P1 = 0.9*(1/3)/Pa
print(f'P1 = {P1}')
P2 = 0.8*(1/3)/Pa
print(f'P2 = {P2}')
P3 = 0.6*(1/3)/Pa
print(f'P3 = {P3}')
print('\n–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–π —Å–ø–æ—Å–æ–±:')
print('–ü–æ—Å–∫–æ–ª—å–∫—É –≤—ã–±–æ—Ä –±–∏–∞—Ç–ª–æ–Ω–∏—Å—Ç–æ–≤ —Ä–∞–≤–Ω–æ–≤–µ—Ä–æ—è—Ç–µ–Ω –Ω–æ—Ä–º–∏—Ä—É–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Çm –ø–æ–ø–∞–¥–∞–Ω–∏—è –æ—Ç–¥–µ–ª—å–Ω–æ–≥–æ –±–∏–∞—Ç–ª–æ–Ω–∏—Å—Ç–∞ –∫ —Å—É–º–º–µ\
–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –ø–æ–ø–∞–¥–∞–Ω–∏—è –∫–∞–∂–¥–æ–≥–æ –±–∏–∞—Ç–ª–æ–Ω–∏—Å—Ç–∞:')
print('Pn = P(ùêµn)/P(ùêµ1)+ùëÉ(ùêµ2)+ùëÉ(ùêµ3)')
P1 = 0.9/(0.9+0.8+0.6)
print(f'P1 = {P1}')
P2 = 0.8/(0.9+0.8+0.6)
print(f'P2 = {P2}')
P2 = 0.6/(0.9+0.8+0.6)
print(f'P3 = {P3}')

```

```python
# –í —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç—ã A –∏ B –ø–æ—Å—Ç—É–ø–∏–ª–æ —Ä–∞–≤–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤, –∞ –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç C —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –ø–æ—Å—Ç—É–ø–∏–ª–æ —Å—Ç–æ–ª—å–∫–æ –∂–µ, 
# —Å–∫–æ–ª—å–∫–æ –Ω–∞ A –∏ B –≤–º–µ—Å—Ç–µ. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ A —Å–¥–∞—Å—Ç –ø–µ—Ä–≤—É—é —Å–µ—Å—Å–∏—é, —Ä–∞–≤–Ω–∞ 0.8. –î–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ 
# B —ç—Ç–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ä–∞–≤–Ω–∞ 0.7, –∞ –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ C - 0.9. –°—Ç—É–¥–µ–Ω—Ç —Å–¥–∞–ª –ø–µ—Ä–≤—É—é —Å–µ—Å—Å–∏—é. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –æ–Ω —É—á–∏—Ç—Å—è: 
# a). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ A –±). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ B –≤). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ C?

print('–û–±–æ–∑–Ω–∞—á–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ  —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ A = x, —Ç–æ–≥–¥–∞ B = x, C = 2x')
print('–ê–ø—Ä–∏–æ—Ä–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ P(B)a = x/4x = 1/4, P(B)b = x/4x =1/4, P(B)c = 2x/4x = 1/2 ')
print('–ó–∞–ø–∏—à–µ–º —Ñ–æ—Ä–º—É–ª—É –ë–∞–π–µ—Å–∞ –¥–ª—è –ø—Ä–æ–∏–∑–æ—à–µ–¥—à–µ–≥–æ —Å–æ–±—ã—Ç–∏—è: P(B|A) = P(A|B)*P(B)/P(A)')
print('–í—ã—á–∏—Å–ª–∏–º –ø–æ–ª–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(A) = P(ùêµ_1)‚àóùëÉ(ùê¥‚îÇùêµ_1 )+ùëÉ(ùêµ_2 )‚àóùëÉ(ùê¥‚îÇùêµ_2 )+ùëÉ(ùêµ_3 )‚àóùëÉ(ùê¥‚îÇùêµ_3 )')
Pa = (1/4)*(0.8+0.7+2*0.9)
print(f'Pa = {Pa}')
print('–í—ã—á–∏—Å–ª–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ –ê, B, C')
PfakA = 0.8*(1/4)/Pa
print(f'P(B|A)a = {PfakA}')
PfakB = 0.7*(1/4)/Pa
print(f'P(B|A)b = {PfakB}')
PfakC = 0.9*(1/2)/Pa
print(f'P(B|A)c = {PfakC}')

```

```python
# –£—Å—Ç—Ä–æ–π—Å—Ç–≤–æ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ —Ç—Ä–µ—Ö –¥–µ—Ç–∞–ª–µ–π. –î–ª—è –ø–µ—Ä–≤–æ–π –¥–µ—Ç–∞–ª–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã–π—Ç–∏ –∏–∑ —Å—Ç—Ä–æ—è –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü —Ä–∞–≤–Ω–∞ 0.1, –¥–ª—è –≤—Ç–æ—Ä–æ–π - 0.2,
# –¥–ª—è —Ç—Ä–µ—Ç—å–µ–π - 0.25. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü –≤—ã–π–¥—É—Ç –∏–∑ —Å—Ç—Ä–æ—è: –∞). –≤—Å–µ –¥–µ—Ç–∞–ª–∏ –±). —Ç–æ–ª—å–∫–æ –¥–≤–µ –¥–µ—Ç–∞–ª–∏ –≤). 
# —Ö–æ—Ç—è –±—ã –æ–¥–Ω–∞ –¥–µ—Ç–∞–ª—å –≥). –æ—Ç –æ–¥–Ω–æ–π –¥–æ –¥–≤—É—Ö –¥–µ—Ç–∞–ª–µ–π?
print('A)')
print('–ü–æ—Å—á–∏—Ç–∞–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ö–æ–¥–∞ –∏–∑ —Å—Ç—Ä–æ—è 3 –¥–µ—Ç–∞–ª–µ–π –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü –∫–∞–∫ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å 3-—Ö –Ω–µ–∑–∞–≤–∏—Å —Å–æ–±—ã—Ç–∏–π P(123) = P1*P2*P3')
P123 = 0.1*0.2*0.25
print(f'P–∞ = {P123}')
print('–ë)')
print('–ü–æ—Å—á–∏—Ç–∞–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ö–æ–¥–∞ –∏–∑ —Å—Ç—Ä–æ—è 2 –¥–µ—Ç–∞–ª–µ–π –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü, –∫–∞–∫ —Å—É–º–º—É –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π 3-—Ö –Ω–µ–∑–∞–≤–∏—Å—ã–º—ã—Ö —Å–æ–±—ã—Ç–∏–π\n \
1-—Å–ª–æ–º–∞–ª–∞—Å—å 2- —Å–ª–æ–º–∞–ª–∞—å 3 - –Ω–µ —Å–ª–æ–º–∞–ª–∞—Å—å P3–Ω = 1-P3. P–± = P1*P2*P3–Ω+P1*P2–Ω*P3+P1–Ω*P2*P3')
Pb = 0.1*0.2*0.75 + 0.1*0.8*0.25 + 0.9*0.2*0.25
# Pb = 0.1*0.2 + 0.1*0.25 + 0.2*0.25
print(f'P–± = {Pb}')
print('–í)')
print('–ü–æ—Å—á–∏—Ç–∞–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ö–æ–¥–∞ –∏–∑ —Å—Ç—Ä–æ—è —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ–π –¥–µ—Ç–∞–ª–∏, –¢.–µ. –æ–¥–Ω–æ–π –¥–≤—É—Ö –∏–ª–∏ —Ç—Ä–µ—Ö. –ü–æ—Å–ª–µ–¥–Ω–∏–µ 2 –Ω–∞–º –∏–∑–≤–µ—Å—Ç–Ω—ã \
—Ä–∞—Å—á–∏—Ç–∞–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ö–æ–¥–∞ –∏–∑ —Å—Ç—Ä–æ—è 1 –¥–µ—Ç–∞–ª–∏. ')
Pc1 = 0.2*0.9*0.75+0.1*0.8*0.75+0.25*0.9*0.8
Pc = P123 + Pb + Pc1
print(f'P–≤ = {Pc}')
print('–ì)')
print('–ü–æ—Å—á–∏—Ç–∞–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Ö–æ–¥–∞ –∏–∑ —Å—Ç—Ä–æ—è —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ–π –∏–ª–∏ –¥–≤—É—Ö, –∫–∞–∫ —Å—É–º–º—É –≤—ã—Ö–æ–¥–∞ –æ–¥–Ω–æ–π –∏ –¥–≤—É—Ö –¥–µ—Ç–∞–ª–µ–π')
Pg = Pb + Pc1
print(f'P–≥ = {Pg}')
# p = P123 + Pb + Pc + Pg
# print(p)

```
